# HEAT_MAP

```python
# everythin mas be the same in units
# correct the data frame 
# set the correct index

df = df.set_index("Countries")
```

```python
df.head(1)

# drop life expecrancy
# so that all columns are the same 
```

```python
df = df.drop("Life expectancy", axis=1)
```

```python
df.head()
```

```python
plt.figure(figsize=(9,9),dpi=300)

sns.heatmap(df);
```

```python
plt.figure(figsize=(9,9),dpi=300)

sns.heatmap(df,linewidth= 0.5);
# to separate the tables
```

```python
# use to place the values in the tales

plt.figure(figsize=(9,9),dpi=300)

sns.heatmap(df,linewidth= 0.5, annot =True,
           cmap="viridis"); # take a note that 
# the cmap changes spectrum
# not all spectrums are visable for color blinded people
# defaul colors are set for color blind 
```

```python
# may center the color bar 

plt.figure(figsize=(9,9),dpi=300)

sns.heatmap(df,linewidth= 0.5, annot =True,
           cmap="viridis", center = 40);
# that is how to effect outliers
```

# CLUSTER_MAP

```python
# to group the columns and rows

plt.figure(figsize=(9,9),dpi=300)

sns.clustermap(df,linewidth= 0.5, annot =True,
           cmap="viridis"); # use col_cluster = False
# to turn off column c

# look at column 
# europe is clustered with North america
# since their values are alined 

# cluster by columns
# Africa has birth rate and infinate mortalaity rate 
# in the same category
```
