# Scatter Plots

```python
sns.scatterplot(x="salary",y="sales", data=df);

# call function from seaborn libriary
```

```python
# parameters adjustmnet comes from matplotlib
# before you even call scatter
# you call matplotlib first
# next seaborn adjust to those parameters

plt.figure(figsize=(12,4), dpi=300)
sns.scatterplot(x="salary",y="sales", data=df);
```

```python
# to color the dots use hue= command
# by default equal to false

plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df, hue="level of education");

# it automatically colors the points 
# and gives legends
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df, hue="sales");

# if you choose contineous will become gradient 
# able to show 3 dimentions of information on 2d plot
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df, hue="level of education",
               palette="Dark2");

# to change colors use colormap 
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df, size="work experience");

# able to make plots bigger 
# use salary as a term of destiguesher 
# to make points bigger
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df,
               s=200);
# this will make points more larger
```

```python
# if you have point on top of point 
# use alpha to make the data transperent 
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df,alpha=0.5, s=200);
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df, s=200,
               style="level of education", hue="level of education");
# make it easier to read or
# may add another layer to it
```

```python
# save a file
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary",y="sales", data=df, s=200,style="level of education", hue="level of education")

plt.savefig("my_plot.jpg")
```
