# CAT plot

```python
sns.catplot(data=df, x="gender", y="math score");

# statistical distribution within categories
```

```python
plt.figure(dpi=300)

sns.catplot(data=df, x="gender", y="math score", kind="box",
           row="lunch");

# now it will take above graph
# and will create rows for every category of lunch
# similar to hue command that makes everythin in one graph
```

```python
plt.figure(dpi=300)

sns.catplot(data=df, x="gender", y="math score", kind="box",
           col="lunch", );

# now lets add another parameter, the same as above
# but in a form of colums
```

# PAIR grid

```python
g =sns.PairGrid(df)

# will create an empty boxes based on the columns 
# waiting for the mapping commands

# the columns are chosen automatically 
# as in pairplot (pairplot is the run of grid)
# all numeric columns are run on each other
g
```

```python
g =sns.PairGrid(df)
# choose what graph to use in
#### upper diagnal 

g = g.map_upper(sns.scatterplot)
```

```python
g =sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

# for lower diagnal
g = g.map_lower(sns.kdeplot)
```

```python
g =sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)

# midle diagnal
# remember, they have to make sence
# since it will place the variabel on the variable itself
g = g.map_diag(sns.kdeplot)
```

```python
g =sns.PairGrid(df, hue="gender")
# make more use of hue to double the information

g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)

# midle diagnal
# remember, they have to make sence
# since it will place the variabel on the variable itself
g = g.map_diag(sns.kdeplot)

# add legend
g = g.add_legend()
```
