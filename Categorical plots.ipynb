# Count plot

# Bar plot
allows to choose which metrix to count on

```python
plt.figure(figsize=(10,5), dpi = 200) 
sns.barplot(data=df,x ="level of education", y = "salary",
           estimator=np.mean, # aggregation to display in plot
           ci = "sd",hue="division"); # black line = std  

# division has no std because only 3 evaluations are avaliable
plt.legend(loc=(1.1,0.5))
plt.show()
```

# DISTRIBUTION WITH CATEGORIES

```python
df = pd.read_csv("/Users/vazgenosipov/Desktop/Python/Python Data analys/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv")

```

```python
df.head()
```

# BOX plot

```python
sns.boxplot(data=df,y="math score");

# for all studetns and all conditions
```

```python
plt.figure(figsize=(10,5), dpi=200) # add figure parameters

sns.boxplot(data=df,y="reading score", x = "parental level of education");

# this will seperate the graphs by groups
```

### Create a sub category

```python
plt.figure(figsize=(10,5), dpi=200)
sns.boxplot(data=df,y="reading score", x = "parental level of education",
           hue="test preparation course");

plt.legend(bbox_to_anchor=(1.1,0.5))

# hue will devide the scores into sub categories 
# it is used to break the graph into colors
```

### horizontal box plot

```python
plt.figure(figsize=(8,6), dpi=200)
sns.boxplot(data=df,x="reading score", y = "parental level of education",
           hue="test preparation course", palette="Set1");
# just switch x and y

plt.legend(bbox_to_anchor=(0.5,-0.1))
```

# VIOLIN Plot

```python
plt.figure(figsize=(8,7), dpi=200)
sns.violinplot(data=df,x="reading score");

# shows kde with a box plon inside and a dot at center
```

```python
plt.figure(figsize=(8,7), dpi=200)
sns.violinplot(data=df,x="reading score", y = "parental level of education");

# may make multiple with hue
# a violin is a mirrored kde plot
```

```python
plt.figure(figsize=(8,10), dpi=200)
sns.violinplot(data=df,x="reading score", y = "parental level of education",
           hue="test preparation course", palette="Set2")

plt.legend(bbox_to_anchor=(0.5,-0.1));

# make make sub categories with hue
```

```python
# to make subcategoris more readable 
# use command split=True

plt.figure(figsize=(8,10), dpi=200)
sns.violinplot(data=df,x="reading score", y = "parental level of education",
           hue="test preparation course", split=True,
              inner="quartile")

### inner = none will remove the box plot inside
# inner = 'quartile' will mark quartilers

# one hals is one catetogy 
# second half is another

plt.legend(bbox_to_anchor=(0.5,-0.1));
```

```python
plt.figure(figsize=(8,10), dpi=200)
sns.violinplot(data=df,x="reading score", y = "parental level of education",
              inner="stick",
              bw=0.1);

# bw = cleares the noice and show the data centration points 
# inner = 'stick'
# will show every instance of the data
```

# SWARM plot

```python
plt.figure(figsize=(8,5), dpi = 200)

# make dots smaller
sns.swarmplot(data=df, x="math score", size=5);
```

```python
plt.figure(figsize=(8,5), dpi = 200)

# make categoris
sns.swarmplot(data=df, x="math score",y = "gender", size=3);
```

```python
plt.figure(figsize=(10,6), dpi = 200)

# make sub-categoris
sns.swarmplot(data=df, x="math score",y = "gender", size=5,
             hue='test preparation course') # not informative

plt.legend(bbox_to_anchor=(0.5,-0.1));
```

```python
plt.figure(figsize=(10,6), dpi = 200)

# make sub-categoris, split them
sns.swarmplot(data=df, x="math score",y = "gender", size=5,
             hue='test preparation course',dodge=True) 

plt.legend(bbox_to_anchor=(0.5,-0.1));
```

# BOXEN plot

```python
plt.figure(figsize=(10,6), dpi = 200)

sns.boxenplot(data=df,x="math score", y="test preparation course")
```
