# JOIN plots

```python
plt.figure(figsize=(6,5),dpi = 200)

sns.jointplot(data=df, x="math score", y="reading score",
             alpha=0.3); # make the scatter a bit trasperent
# by defail the insode plot is scatter
```

```python
# may change the insde plot by kind command 

plt.figure(figsize=(6,5),dpi = 200)

sns.jointplot(data=df, x="math score", y="reading score", 
              kind = "hex");

# the more point fall into hexagons the darket it becomes
```

```python
# same as above but now it matches with the blocks on the sides

plt.figure(figsize=(6,5),dpi = 200)

sns.jointplot(data=df, x="math score", y="reading score", 
              kind = "hist");
```

```python
# may create a shaded KDE plot to show the density
# note that it has changed the sides to kde

plt.figure(figsize=(6,5),dpi = 200)

sns.jointplot(data=df, x="math score", y="reading score", 
              kind = "kde", shade=True)
```

```python
### DISPLAY multiple conditions
# with hue 

plt.figure(figsize=(6,5),dpi = 200)

sns.jointplot(data=df, x="math score", y="reading score",
              hue="gender", palette="Set1")
```
