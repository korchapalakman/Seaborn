## #1 Create a rug plot

```python
plt.figure(figsize=(5,3),dpi=200)
sns.rugplot(x="salary", data=df, height=0.5)

# increase the size of the ticks to make them visible
# shown in percent of the total image
plt.tight_layout();
```

## #2 Create Distribution plot (displot)
distribution plot

describe the number of bins

```python
sns.set(style="darkgrid") # gives a dark background
sns.displot(data=df,x="salary", bins=20); # number of bins
```

```python
sns.set(style="whitegrid") # gives a white background
sns.displot(data=df,x="salary", bins=20);
```

```python
sns.set(style="ticks") # gives tiks along y axes
# this is a high-level command that will style all of your notebook
# may add rug to it

sns.displot(data=df,x="salary", bins=20, rug=True);
```

```python
sns.displot(data=df,x="salary", bins=20, color="red", edgecolor="black",
           linewidth=1.5,ls="-.");

# we are able to use matplotlib inside sns
# since it has kwargs in the attributes - key words attributs
```

### Histplot

```python
sns.histplot(data=df,x="salary",kde=True, 
             color="green",edgecolor="black",lw=1.5);

# histogramm
# may use KDE on top of 
```

## Example 2

```python
# create random data with seed

np.random.seed(42)
sample_sages = np.random.randint(0,100,200)
```

```python
sample_sages
# array
```

```python
# make a dataframe

sample_sages = pd.DataFrame(sample_sages, columns = ["age"])
```

```python
sample_sages
```

```python
# create rug plot

sns.rugplot(data=sample_sages, x = "age", height=0.5);
```

```python
# create distribution 

sns.set(style="white")
sns.displot(data=sample_sages, x = "age", rug=True
            ,edgecolor="black", bins=10,kde=True);

# as you see kde is cut off because there are no values
```

```python
sns.kdeplot(data=sample_sages, x = "age")

# as you see here you get unecessary range of values that dont exist
```

```python
sns.kdeplot(data=sample_sages, x = "age", clip=[0,100])

# the clip command will cut off
# the unecessary data
```

```python
# to pick the variation 
# use bw_adjust
# uses the % to evaluete the noice

sns.kdeplot(data=sample_sages, x = "age", clip=[0,100], bw_adjust=0.1)
```

```python
sns.kdeplot(data=sample_sages, x = "age", clip=[0,100], 
           shade=True, color="r")

# use shade to set the background
```
